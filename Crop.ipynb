```python
from skimage import img_as_ubyte

for i in range (len(gts)):
    gts[i] = img_as_ubyte(gts[i])
    gts[i] = color.rgb2gray(gts[i])
    
    
    
print(gts[0][2000][3326])
```

```python
from skimage.measure import label
from skimage.measure import regionprops

w = 454
h = 454

for i in range (len(gts)):
    regions_label, num_labels = label(gts[i], connectivity = 2, return_num = True)
    props = regionprops(regions_label)
#    plt.imshow(regions_label)
#    plt.show()
    name = names_gt[i]
    filename_img = "C:\\Study\\2sem\\Java\\Project\\dataset\\images\\" + name + ".tif"
    img = cv2.imread(filename_img)
    m = 0
    print(num_labels)
    for i in range (num_labels):
        center = props[i].centroid
        r0 = int(center[0] - int(h/2))
        r1 = int(center[0] + int(h/2))
        c0 = int(center[1] - int(h/2))
        c1 = int(center[1] + int(h/2))
        if (r0 < 0):
            r0 = 0
            r1 = h
        if (r1 > img.shape[0]):
            r0 = img.shape[0]-h
            r1 = img.shape[0]
        if (c0 < 0):
            c0 = 0
            c1 = w
        if (c1 > img.shape[1]):
            c0 = img.shape[1]-h
            c1 = img.shape[1]
    
        mass = img[r0:r1,c0:c1]
        print(mass.shape)
        cv2.imwrite(".\\mass\\" + name + "_" + str(m) + ".png", mass)
        
        m = m+1
#         plt.imshow(mass)
#         plt.show()
    
    
```

```python
from skimage.measure import label
from skimage.measure import regionprops

w = 454
h = 454

for i in range (len(gts)):
    regions_label, num_labels = label(gts[i], connectivity = 2, return_num = True)
    props = regionprops(regions_label)
#    plt.imshow(regions_label)
#    plt.show()
    name = names_gt[i]
    filename_img = "C:\\Study\\2sem\\Java\\Project\\dataset\\images\\" + name + ".tif"
    img = cv2.imread(filename_img)
    print(num_labels)
    for i in range (num_labels):
        b = props[i].bbox
        print(b)
        r_min = b[0]
        r_max = b[3]
        c_min = b[1]
        c_max = b[4]
        if (r_min + 454 < img.shape[0]+1) and (c_min + 454 < img.shape[1]+1):
            mass = img[r_min:r_min+454,c_min:c_min+454]
            print(mass.shape)
            cv2.imwrite(".\\mass2\\" + name + "_00.png", mass)
        if (r_min + 454 < img.shape[0]+1) and (c_max - 454 > 0):
            mass = img[r_min:r_min+454,c_max - 454:c_max]
            print(mass.shape)
            cv2.imwrite(".\\mass2\\" + name + "_01.png", mass)
        if (r_max - 454 > 0) and (c_min + 454 < img.shape[1]+1):
            mass = img[r_max - 454:r_max,c_min:c_min +454]
            print(mass.shape)
            cv2.imwrite(".\\mass2\\" + name + "_10.png", mass)
        if (r_max - 454 > 0) and (c_max - 454 > 0):
            mass = img[r_max - 454:r_max,c_max - 454:c_max]
            print(mass.shape)
            cv2.imwrite(".\\mass2\\" + name + "_11.png", mass)
    
        
        
        
#         plt.imshow(mass)
#         plt.show()
```

```python
from skimage.transform import rotate

massfiles = sorted(glob.glob("C:\\Study\\2sem\\Java\\Project\\mass\\*.png"))
images = [cv2.imread(file) for file in massfiles]

names = [os.path.splitext(os.path.basename(file))[0] for file in massfiles]
print(names[0])

for i in range (len(images)):
    name = names[i]
    image = images[i]
    
    rows,cols, color = image.shape
     
    M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
    dst = cv2.warpAffine(image,M,(cols,rows))
    
    cv2.imwrite(".\\massRot\\" + name + "_r1.png", dst)
    
    dst = cv2.warpAffine(dst,M,(cols,rows))
    cv2.imwrite(".\\massRot\\" + name + "_r2.png", dst)
    
    dst = cv2.warpAffine(dst,M,(cols,rows))
    cv2.imwrite(".\\massRot\\" + name + "_r3.png", dst)
    
```
